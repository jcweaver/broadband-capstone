```python
import pandas as pd
import numpy as np

```

```python
distance = pd.read_csv("../joined_wDistance.csv", index_col=0)

acs = pd.read_csv("acs_demographic_data.csv", index_col=0)
```

```python
#Some cleanup to remove unnecessary columns &rename geoid to match casing
distance = distance.drop(columns = ['Unnamed: 0.1','Unnamed: 0.1.1' ])
distance.rename(columns={'Geoid' : 'GEOID'}, inplace=True)
```

```python
distance
```

```python
acs
```

## Merging the Data

```python
merged = distance.merge(acs, how = 'left', on = 'GEOID')
merged
```

```python
merged.to_csv("../merged_data.csv", index=False)
```
