## Additional Data Merging

Adding additional data to the test/train set and creating larger groupings of age/income stats.

```python
def update_tract(x):
    tract = str(x)
    if len(tract) < 11:
        tract = "0" + tract
    return tract
```

```python
train = pd.read_csv("../data/broadband_training_bin_1.csv", converters={"tract_geoid" : lambda x: update_tract(x),
                                                                       'All_Providers' : lambda x: x.replace("{","").replace("}", "").split(", ")})
train.head()
```

```python
test = pd.read_csv("../data/broadband_test_bin_1.csv", converters={"tract_geoid" : lambda x: update_tract(x),
                                                                  'All_Providers' : lambda x: x.replace("{","").replace("}", "").split(", ")})
test.head()
```

```python
train.shape
```

```python
test.shape
```

```python
ddi = pd.read_excel("../data/DDI_export_2019 - Copy.xlsx", sheet_name="US Tracts", converters= {"FIPS" : lambda x: str(x)})
ddi.head()
```

```python
## Step 1: Merge DDI Data

new_train = train.merge(ddi[["FIPS", "SE", "INFA", "DDI"]], how="left", left_on="tract_geoid", right_on = "FIPS")
new_train.head()
```

```python
new_train.isnull().sum()
```

```python
## Now one row is missing. let's drop it

new_train[new_train.FIPS.isna()][["tract_geoid", "total_population"]]
```

```python
new_train = new_train[~new_train.FIPS.isna()]
```

```python
new_test = test.merge(ddi[["FIPS", "SE", "INFA", "DDI"]], how="left", left_on="tract_geoid", right_on = "FIPS")
new_test.head()
```

```python
new_test.isnull().sum()
```

```python
## Step 2: Make bigger buckets of income and age
pd.set_option('display.max_rows', None)
new_train.columns.to_list()
```

```python
## Add additional columns for incomes

new_train['pct_pop_income_lt_50k'] = (new_train['pct_pop_lt_10k'] + new_train['pct_pop_10k_thru_15k'] +
                                  new_train['pct_pop_15k_thru_20k'] + new_train['pct_pop_20k_thru_25k'] + 
                                  new_train['pct_pop_25k_thru_30k'] + new_train['pct_pop_30k_thru_35k'] +
                                  new_train['pct_pop_35k_thru_40k'] + new_train['pct_pop_40k_thru_45k'] +
                                  new_train['pct_pop_45k_thru_50k']
                                 )

new_train['pct_pop_income_lt_30k'] = (new_train['pct_pop_lt_10k'] + new_train['pct_pop_10k_thru_15k'] +
                                  new_train['pct_pop_15k_thru_20k'] + new_train['pct_pop_20k_thru_25k'] + 
                                  new_train['pct_pop_25k_thru_30k']
                                 )

new_train['pct_pop_income_gt_100k'] = (new_train['pct_pop_100k_thru_125k'] + new_train['pct_pop_125k_thru_150k'] +
                                   new_train['pct_pop_150k_thru_200k'] + new_train['pct_pop_gt_200k']
                                 )

new_test['pct_pop_income_lt_50k'] = (new_test['pct_pop_lt_10k'] + new_test['pct_pop_10k_thru_15k'] +
                                  new_test['pct_pop_15k_thru_20k'] + new_test['pct_pop_20k_thru_25k'] + 
                                  new_test['pct_pop_25k_thru_30k'] + new_test['pct_pop_30k_thru_35k'] +
                                  new_test['pct_pop_35k_thru_40k'] + new_test['pct_pop_40k_thru_45k'] +
                                  new_test['pct_pop_45k_thru_50k']
                                 )

new_test['pct_pop_income_lt_30k'] = (new_test['pct_pop_lt_10k'] + new_test['pct_pop_10k_thru_15k'] +
                                  new_test['pct_pop_15k_thru_20k'] + new_test['pct_pop_20k_thru_25k'] + 
                                  new_test['pct_pop_25k_thru_30k']
                                 )

new_test['pct_pop_income_gt_100k'] = (new_test['pct_pop_100k_thru_125k'] + new_test['pct_pop_125k_thru_150k'] +
                                   new_test['pct_pop_150k_thru_200k'] + new_test['pct_pop_gt_200k'])
```

```python
## Add additional columns for ages

new_train["pct_ages_gt_50"] = (new_train['pct_pop_50_to_54'] + new_train['pct_pop_55_to_59'] + new_train['pct_pop_60_to_64'] +
                           new_train['pct_pop_65_to_69'] + new_train['pct_pop_70_to_74'] + new_train['pct_pop_75_to_79'] + 
                           new_train['pct_pop_80_to_84'] + new_train['pct_pop_gt_85']
                                     )

new_train["pct_ages_lt_19"] = (new_train['pct_pop_lt_5'] + new_train['pct_pop_5_to_9']+ new_train['pct_pop_10_to_14']+
                               new_train['pct_pop_15_to_19'] )


new_test["pct_ages_gt_50"] = (new_test['pct_pop_50_to_54'] + new_test['pct_pop_55_to_59'] + new_test['pct_pop_60_to_64'] +
                           new_test['pct_pop_65_to_69'] + new_test['pct_pop_70_to_74'] + new_test['pct_pop_75_to_79'] + 
                           new_test['pct_pop_80_to_84'] + new_test['pct_pop_gt_85']
                                     )

new_test["pct_ages_lt_19"] = (new_test['pct_pop_lt_5'] + new_test['pct_pop_5_to_9']+ new_test['pct_pop_10_to_14']+ 
                              new_test['pct_pop_15_to_19'])
```

```python
## Step 3: RUCA aggregation

new_train["ruca_metro"] = new_train["Primary RUCA Code 2010"].apply(lambda x: x<=3)
new_train["ruca_micro"] = new_train["Primary RUCA Code 2010"].apply(lambda x: x>=4 and x<=6)
new_train["ruca_small_town"] = new_train["Primary RUCA Code 2010"].apply(lambda x: x>=7 and x<=9)
new_train["ruca_rural"] = new_train["Primary RUCA Code 2010"].apply(lambda x: x==10)


new_test["ruca_metro"] = new_test["Primary RUCA Code 2010"].apply(lambda x: x<=3)
new_test["ruca_micro"] = new_test["Primary RUCA Code 2010"].apply(lambda x: x>=4 and x<=6)
new_test["ruca_small_town"] = new_test["Primary RUCA Code 2010"].apply(lambda x: x>=7 and x<=9)
new_test["ruca_rural"] = new_test["Primary RUCA Code 2010"].apply(lambda x: x==10)
```

```python
## Step 4: Add booleans for big providers

def check_for_att(row):
    if len(row) > 0:
        for i in row:
            #print(i)
            if int(i) > 67052:
                return True
    return False
```

```python

new_train["Comcast_present"] = new_train["All_Providers"].apply(lambda row: '55574' in row)

new_test["Comcast_present"] = new_test["All_Providers"].apply(lambda row: '55574' in row)

new_train["ATT_present"] = new_train["All_Providers"].apply(check_for_att)

new_test["ATT_present"] = new_test["All_Providers"].apply(check_for_att)

new_train["HughesNet_present"] = new_train["All_Providers"].apply(lambda row: '54694' in row)

new_test["HughesNet_present"] = new_test["All_Providers"].apply(lambda row: '54694' in row)

new_train["GCI_Comm_Corp_present"] = new_train["All_Providers"].apply(lambda row: '59349' in row)

new_test["GCI_Comm_Corp_present"] = new_test["All_Providers"].apply(lambda row: '59349' in row)

new_train["ViaSat_present"] = new_train["All_Providers"].apply(lambda row: '56004' in row)

new_test["ViaSat_present"] = new_test["All_Providers"].apply(lambda row: '56004' in row)

new_train["VSAT_present"] = new_train["All_Providers"].apply(lambda row: '55262' in row)

new_test["VSAT_present"] = new_test["All_Providers"].apply(lambda row: '55262' in row)

new_train["Century_Link_present"] = new_train["All_Providers"].apply(lambda row: '53788' in row)

new_test["Century_Link_present"] = new_test["All_Providers"].apply(lambda row: '53788' in row)

new_train["Spectrum_present"] = new_train["All_Providers"].apply(lambda row: '58623' in row)

new_test["Spectrum_present"] = new_test["All_Providers"].apply(lambda row: '58623' in row)

new_train["Crown_Castle_present"] = new_train["All_Providers"].apply(lambda row: '55027' in row)

new_test["Crown_Castle_present"] = new_test["All_Providers"].apply(lambda row: '55027' in row)

new_train["Etheric_present"] = new_train["All_Providers"].apply(lambda row: '57777' in row)

new_test["Etheric_present"] = new_test["All_Providers"].apply(lambda row: '57777' in row)

new_train["Frontier_Communications_present"] = new_train["All_Providers"].apply(lambda row: ('59351' in row) or 
                                                                               ('59347' in row))

new_test["Frontier_Communications_present"] = new_test["All_Providers"].apply(lambda row: ('59351' in row) or 
                                                                               ('59347' in row))

```

```python
new_train.columns.to_list()
```

```python
## Remove some columns

new_train = new_train.drop(columns=["FIPS", "All_Providers", "AllMaxAdDown", "AllMaxAdUp"])
new_test = new_test.drop(columns=["FIPS", "All_Providers", "AllMaxAdDown", "AllMaxAdUp"])
```

```python
new_train = new_train.replace(True, 1)
new_train = new_train.replace(False, 0)

new_test = new_test.replace(True, 1)
new_test = new_test.replace(False, 0)
```

```python
## Reading in the data

train = pd.read_csv("../data/train_3.csv")
test = pd.read_csv("../data/test_3.csv")

new_data = pd.read_csv("../census_scripts/census_computing_devices.csv", index_col="Unnamed: 0")
```

```python
train = train.drop(columns=['NAME_y'])
train = train.rename(columns= {'NAME_x' : 'NAME', 'state_x' : 'state',
 'county_x' : 'county',
 'tract_x' : 'tract'})
train['pct_of_internet_users_with_broadband'] = train['pct_internet_broadband_any_type'] / train['pct_internet']*100
train['log_median_income_over_log_median_house'] = train['log_median_income'] / train['log_median_house_value']

## Next replace "computer" in column names with computing_device
train.columns = train.columns.str.replace("computer", "computing_device")
```

```python
pd.set_option('display.max_rows', None)
train.columns.to_list()
```

```python
test = test.drop(columns=['NAME_y'])
test = test.rename(columns= {'NAME_x' : 'NAME', 'state_x' : 'state',
 'county_x' : 'county',
 'tract_x' : 'tract'})
test['pct_of_internet_users_with_broadband'] = test['pct_internet_broadband_any_type'] / test['pct_internet']*100
test['log_median_income_over_log_median_house'] = test['log_median_income'] / test['log_median_house_value']

## Next replace "computer" in column names with computing_device
test.columns = test.columns.str.replace("computer", "computing_device")
```

```python
def safe_division(x, y):
    try:
        return (x / y)*100
    except:
        #y would have been 0 so return NAN
        return np.nan
```
