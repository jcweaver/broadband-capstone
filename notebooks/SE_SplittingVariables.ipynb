```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/home/ec2-user/efs/fcc_census_3_augmented.csv')
```

## Splitting into One-Hot Encoded Columns

### Provider Counts

```python
data['0_5_providers'] =  (data['All_Provider_Count']<=5).astype(int)
data['6_10_providers'] = ((data['All_Provider_Count']>5) & (data['All_Provider_Count']<=10)).astype(int)
data['11_15_providers'] = ((data['All_Provider_Count']>11) & (data['All_Provider_Count']<=15)).astype(int)
data['16_20_providers'] = ((data['All_Provider_Count']>16) & (data['All_Provider_Count']<=20)).astype(int)
data['20+_providers'] = (data['All_Provider_Count']>20).astype(int)
```

```python
data[['All_Provider_Count','0_5_providers','6_10_providers','11_15_providers','16_20_providers','20+_providers']]
```

```python
data['single_provider'] = (data['All_Provider_Count']==1).astype(int)
```

### Broadband Speeds

```python
speeds = [0, 25, 99, 499, np.inf]
speed_columns = ['MaxAdDown',
                 'Ookla Median Download Speed (Mbps)', 
                 'Form 477 All Terrestrial Broadband: Max Advertised Consumer Download Speed (Mbps)']

for speed_type in speed_columns:
    for i in range(len(speeds)-1):
        data[str(speeds[i]) + "_" + str(speeds[i+1]) + "_" + speed_type] = ((data[speed_type]>speeds[i]) & (data[speed_type]<=speeds[i+1])).astype(int)
```

```python
data[['MaxAdDown','0_25_MaxAdDown','25_99_MaxAdDown','99_499_MaxAdDown','499_inf_MaxAdDown']]
```

```python
data[['Form 477 All Terrestrial Broadband: Max Advertised Consumer Download Speed (Mbps)',
      '0_25_Form 477 All Terrestrial Broadband: Max Advertised Consumer Download Speed (Mbps)',
      '25_99_Form 477 All Terrestrial Broadband: Max Advertised Consumer Download Speed (Mbps)',
      '99_499_Form 477 All Terrestrial Broadband: Max Advertised Consumer Download Speed (Mbps)',
      '499_inf_Form 477 All Terrestrial Broadband: Max Advertised Consumer Download Speed (Mbps)']]

```

```python
data[['Ookla Median Download Speed (Mbps)',
      '0_25_Ookla Median Download Speed (Mbps)',
      '25_99_Ookla Median Download Speed (Mbps)',
      '99_499_Ookla Median Download Speed (Mbps)',
      '499_inf_Ookla Median Download Speed (Mbps)']]
```

### RUCA


```python
data[['Primary RUCA Code 2010','Secondary RUCA Code, 2010 (see errata)']]
```

```python
for i in data['Primary RUCA Code 2010'].value_counts().index:
    data['Primary RUCA Code - ' + str(i)] = (data['Primary RUCA Code 2010'] == i).astype(int)

for i in data['Secondary RUCA Code, 2010 (see errata)'].value_counts().index:
    data['Secondary RUCA Code - ' + str(i)] = (data['Secondary RUCA Code, 2010 (see errata)'] == i).astype(int)

```
