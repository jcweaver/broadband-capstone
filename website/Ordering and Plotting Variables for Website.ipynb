```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
training_data = pd.read_csv('/home/ec2-user/capstone/broadband-capstone/data/current_data/train_3.csv')
```

```python
correlations = training_data.corr()
```

```python
plt.figure(figsize = (15,8))
ax = correlations['MaxAdUp'].abs().sort_values(ascending=False)[0:10].plot.bar()
ax.set_xticklabels(ax.get_xticklabels(),rotation = 45)
plt.show()
```

```python
correlations_plot = pd.DataFrame() 
correlations_plot['positive'] = correlations['MaxAdUp']>0
correlations_plot['abs_values'] = correlations['MaxAdUp'].abs()
correlations_plot['index'] = correlations.index
plot_df = correlations_plot.sort_values(by='abs_values',ascending=False)[0:10]

plt.figure(figsize = (15,8))
ax = sns.barplot(data=plot_df, x='index', y='abs_values', hue='positive', dodge=False)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 45)
plt.show()
```
